```python
import numpy as np
import pandas as pd
```

```python
cr=pd.read_csv('/Users/macbookpro/Downloads/CreditRisk.csv')
```

```python
cr.head()
```

```python
cr.info()
```

```python
cr.Gender.value_counts()
```

```python
cr.Gender = cr.Gender.fillna('Male')
```

```python
cr.Gender.replace({'Male':0, 'Female':1}, inplace=True)
```

```python
cr.Married.value_counts()
```

```python
cr.Married = cr.Married.fillna('Yes')
```

```python
cr.Married.replace({'Yes':0, 'No':1}, inplace=True)
```

```python
cr.Dependents.value_counts()
```

```python
cr.Dependents = cr.Dependents.fillna(0.0)
```

```python
cr.Self_Employed.value_counts()
```

```python
cr.Self_Employed = cr.Self_Employed.fillna('No')
```

```python
cr.LoanAmount.value_counts()
```

```python
cr.LoanAmount = cr.LoanAmount.fillna(cr.LoanAmount.mean())
```

```python
cr.Loan_Amount_Term.value_counts()
```

```python
cr.Loan_Amount_Term = cr.Loan_Amount_Term.fillna(cr.Loan_Amount_Term.mean())
```

```python
cr.Credit_History.value_counts()
```

```python
cr.Credit_History = cr.Credit_History.fillna(0)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
cr.drop('Loan_ID', axis=1, inplace=True)
```

```python
cr.Dependents.value_counts()
```

```python
cr[cr.select_dtypes(include=['object']).columns] = cr[cr.select_dtypes(include=['object']).columns].apply(le.fit_transform)
```

```python
from sklearn.model_selection import train_test_split
```

```python
cr_train, cr_test = train_test_split(cr, test_size=0.2)
```

```python
cr_train_x = cr_train.iloc[: , 0:-1]
cr_train_y = cr_train.iloc[:, -1]
```

```python
cr_test_x = cr_test.iloc[: , 0:-1]
cr_test_y = cr_test.iloc[:, -1]
```

```python
cr_train_x.shape
```

```python
cr_test_x.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logi_cr = LogisticRegression()
```

```python
logi_cr.fit(cr_train_x, cr_train_y)
```

```python
pred = logi_cr.predict(cr_test_x)
pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cr_tab = confusion_matrix(cr_test_y , pred)
cr_tab
```

```python
cr_tab.diagonal().sum() * 100 / cr_tab.sum()
```

```python
# Accuracy score
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(cr_test_y , pred)
```

```python
# recall score 
from sklearn.metrics import recall_score
```

```python
recall_score(cr_test_y , pred)
```

```python
# Precision
from sklearn.metrics import precision_score
```

```python
precision_score(cr_test_y , pred)
```

```python
# F1ratio
from sklearn.metrics import f1_score
```

```python
f1_score(cr_test_y , pred)
```

```python
#AUROC 
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
pred_prob = logi_cr.predict_proba(cr_test_x)
pred_prob
```

```python
roc_auc_score(cr_test_y, pred)
```

```python
pred_prob[: , 1]
```

```python
fpr, tpr, ther = roc_curve(cr_test_y, pred_prob[:,1])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(fpr, tpr)
```

```python
plt.plot(fpr, tpr, color='r')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('AUROC on credit risk data')
plt.grid()
plt.text(x=0.5, y=0.5, s='AUROC score is 0.70')
```
